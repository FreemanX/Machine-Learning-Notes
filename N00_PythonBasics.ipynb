# NumPy     
Reference: https://numpy.org/doc/stable/reference/index.html 

```python
import numpy as np

a = np.array([1, 2, 3, 4])  # 1 dim array
print(a)
```

```python
b = np.zeros(4)  # 1 dim array of 4 zeros
print(b)
```

```python
c = np.zeros(shape=(3, 4))  # 2 dim array of zeros with shape (3, 4), 3 rows, 4 cols
print(c)
```

```python
d = np.ones(shape=(2, 3, 4))  # 3 dim array of ones with shape (2, 3, 4)
print(d)
```

```python
e = np.eye(5)  # 2 dim (5, 5) identity matrix
print(e)
```

```python
# create 2 dim array manually
f = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(f)
```

```python
# create a 2 dim random int array
g = np.random.randint(low=-5, high=5, size=(6, 10))
print(g)
```

```python
# create a 2 dim empty array
h = np.empty(shape=(3, 4))
print(h)
```

Access a np array

```python
print(g[0, 3])  # access element in row=0, col=3
print(g[:, 2])  # access all elements in col=2
print(g[3, :])  # access all elements in row=3
```

```python
# 'a:b' index access -- [a, b)
print(g[:2, 3:6])  # access elements in rows 0-1, and cols 2-4
print(g[:, ::2])   # access every 2 col
print(g[:, ::3])   # access every 3 col
```

Change value (reassign values)

```python
print(g[0, 3])
g[0, 3] = 100
print(g[0, 3])
```

```python
print(g[:2, 3:6])
# change sub-arrays (be careful with the shape)
g[:2, 3:6] = np.ones(shape=(2, 3))
print(g[:2, 3:6])
```

NumPy array has many functions. The axis argument controls the behavior.

```python
print(g.max())  # max element in g
print(g.max(axis=0))  # max element in each col
print(g.max(axis=1))  # max element in each row
```

```python
print(g.sum())        # sum of elements in g
print(g.sum(axis=0))  # sum of elements in each col
print(g.sum(axis=1))  # sum of elements in each row
```

To perform dot product between vectors use `np.dot()`    
To perform matrix multiplication use `@`

```python
u = np.array([1, 2, 3, 4])
v = np.array([4, 4, -2, -3])
x = np.array([-2, -2, 3])
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
B = np.array([
    [4, 8],
    [9, 9]
])
```

```python
np.dot(u, v)
```

```python
u @ v
```

```python
np.matmul(B, A)
```

```python
B @ A
```

```python
A @ x  # matrix by vector product (returns a row vector)
```

```python
A @ x.reshape(-1, 1)  # matrix by vector product (returns a col vector)
```

```python
B.T @ B  # Transpose of B multiply B
```

```python
B @ B @ B @ B  # B to the power of 4
```

`np.linalg` provides many useful linear algebra tools.

```python
np.linalg.matrix_power(B, 4)  # B to the power of 4
```

```python
np.linalg.matrix_rank(B)  # compute the rank of matrix
```

```python
np.linalg.det(B)  # compute the determinant of B
```

```python
np.linalg.inv(B)  # find the inverse of B
```

NumPy broadcasting: efficiently apply an operation to a large array using a smaller array. Looks strange. But if done right, it runs very efficiently.

```python
# Add the small array to each row of the large array
Large = np.ones((30, 4))
Small = np.array([1, 2, 3, 4])
Large + Small
# Other broadcasting examples
Large * Small  # multiply each row of Large by Small element-wise
Large - Small  # subtract each row of Large by Small
Large / Small  # divide each row of Large by Small element-wise
```

# Matplotlib    
A visualization library.

```python
import matplotlib.pyplot as plt
```

#### Generate a single simple plot of sin(x)

```python
# creates a grid (np.array) from -6 to 6 containing 1000 points
x_range = np.linspace(start=-6, stop=6, num=1000)
# compute sin(x) for every point in x_range
y_sin = np.sin(x_range)
# create plots
plt.plot(x_range, y_sin)
# show the plot
plt.show()
```

#### Generate a single plot with extra features

```python
x_range = np.linspace(start=-6, stop=6, num=1000)
y_sin = np.sin(x_range)
y_cos = np.cos(x_range)

plt.plot(x_range, y_sin, color='blue', label='sin(x)')
plt.plot(x_range, y_cos, color='red', label='cos(x)')

plt.legend(loc='upper right') # create legend
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin vs. cos")
plt.show()
```

#### Plot with styles

```python
plt.plot(x_range, y_sin, color='blue', label='sin(x)',
         lw=3,            # line thickness
         linestyle='--',  # line style (dashed)
         alpha=0.5        # line transparency
         )
plt.plot(x_range, y_cos, color='red', label='cos(x)',
         lw=6, linestyle='-.', alpha=0.8)

plt.legend(loc='upper right')  # create legend
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin vs. cos")
plt.show()
```

#### Adding layers    
`plt.fill_between` function can be used to add color between boundaries.

```python
plt.plot(x_range, y_sin, color='blue', label='sin(x)',
         lw=3,            # line thickness
         linestyle='--',  # line style (dashed)
         )
plt.plot(x_range, y_cos, color='red', label='cos(x)',
         lw=3, linestyle=':')

plt.fill_between(x_range, y1=y_sin, color='green', alpha=0.4)
plt.fill_between(x_range, y1=y_cos, color='orange', alpha=0.4)

plt.legend(loc='upper right')  # create legend
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sin vs. cos")
plt.show()
```

#### Scatter Plots    

```python
np.random.seed(8964)
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.random.normal(loc=0, scale=1.2, size=x.shape[0])
plt.scatter(x, y)
# limit the plotting range
plt.xlim((-1, 20))
plt.ylim((-2, 5))
plt.show()
```

#### Custom Scatter Plots    
`scatter` has many arguments allowing us to customize the plot.

```python
colors = ['red', 'blue', 'green', 'orange', 'yellow', 'gray', 'pink']
plt.scatter(x, y,
            s=200,                               # control size of each point
            marker='*',                          # use stars instead of dots
            c=colors                             # supports plotting each point in different color
            )
# annotate the green point
plt.annotate("largest y-val",                    # text to annotate with
             xy=(3.3,1.9),                       # xy coords of point to be annotated
             xytext=(5,3),                       # location of the annotating text
             
             # optional: use an arrow, can specify arrow properties in a dict:
             arrowprops={'ec':'green',           # arrow color
                         'arrowstyle': '->',     # arrow style
                         'lw': 2,                # thickness of arrow
                        }  
        )

# annotate the pink point
plt.annotate("largest x-val",                    # text to annotate with
             xy=(7.3,-0.6),                        # xy coords of point to be annotated
             xytext=(10,-1),                     # location of the annotating text
             
             # optional: use an arrow, can specify arrow properties in a dict:
             arrowprops={'ec':'pink',           # arrow color
                         'arrowstyle': '<->',    # arrow style
                         'lw': 3,                # thickness of arrow
                        }  
        )

plt.xlim((-1, 20))
plt.ylim((-2, 5))
plt.show()
```

#### Scatter plots from arrays    
Plot a NumPy array `X` with shape `(n, 2)` where the first col is the x coordinates and the second col contains y coordinates.

```python
X = np.random.normal(0, 1, size=(20, 2))

# approach 1: plot each column separately as in previous scatter example:
plt.scatter(X[:, 0], X[:, 1])
plt.title("scatter via standard approach")
plt.show()

# approach 2: use Python array unpacking, we first transpose the array then unpack
plt.scatter(*X.T)
plt.title("scatter via unpacking")
plt.show()
```

#### More plotting and style examples    
`color_map`: automatically create a list of colors.    
`itertools.product`: create all possible combinations of different styles and colors in `m_styles`, `l_styles`, and `color_map`.    
More on this: https://stackoverflow.com/a/43416566/12709772

```python
import itertools
import matplotlib as mpl

N = 8*4+10
l_styles = ['-','--','-.',':']                    # different line styles
m_styles = ['','.','o','^','*']                   # different marker styles 
color_map = mpl.cm.Dark2.colors                    # qualitative colormap

# loop over all possible combinations of styles+colors using itertools.product
for i, (m,l,c) in zip(range(N),itertools.product(m_styles,l_styles, color_map)):
    plt.plot([0,1,2], [0,2*i,2*i], 
             color=c, 
             linestyle=l,
             marker=m,
             label=i)

# we can create the legend outside of the plot like this:
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol=4);
```

#### Create a grid of plots    
Usage of `plt.subplots()`:   
`fig, axes = plt.subplots(nrows, ncols, ...)`, where      
`fig` refer to the global plot,    
`axes` is a NumPy array with each element corresponding to a subplot.

```python
# create fig and axes objects, specifying number of subplots and the size of the figure
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20,10))

axes[0,1].plot(x_range, y_sin, color='blue', label='sin(x)')
axes[0,1].legend(loc='upper right')
axes[0,1].set_title("sin x")                       # slightly different title command for subplot
axes[0,1].set_xlabel("my x")                       # slightly different xlabel command for subplot
axes[0,1].set_ylabel("my y")                       # slightly different ylabel command for subplot
axes[0,1].set_xlim((-2,2))                         # slightly different change range of x's command for subplot
axes[0,1].set_ylim((-3,2))                         # slightly different change range of y's command for subplot

axes[1,2].plot(x_range, y_cos, color='red', label='cos(x)')
axes[1,2].legend(loc='lower right')
axes[1,2].set_title("cos x")
axes[1,2].set_xlabel("another x")
axes[1,2].set_ylabel("another y")

axes[2,3].scatter(x, y)
axes[2,3].set_title("my scatter")

# we can also loop through the grids, e.g. if we want to plot lines on each plot of varying height:
for i, ax in enumerate(axes.flat):
    ax.axhline(y=0.1*i, color='orange')                     # horizontal line
    ax.axvline(x=1-0.1*i, color='purple')                   # vertical line


fig.suptitle("A global title for the entire grid")
fig.tight_layout()                                          # call to global fig object, helps format grids nicely
plt.show()
```

#### Visualize functions in 3D    
Visualizing the following 2D function:    
$$
f(x,y)=cos(\sqrt{x^2+y^2})
$$
`np.linspace` creates a grid of points -- 1D grid. Then we need to use `np.meshgrid` to create a 2D grid.

```python
def f(x, y):
    return np.cos(np.sqrt(x**2 + y**2))


# Create two 1D grids using linspace
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

# Combine x, y into one 2D grid
X, Y = np.meshgrid(x, y)

# Calculate the value of f(x, y) 
Z = f(X, Y)

# Plot
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```
